# Проект: Построение дашбордов в Tableau

Цели проекта:
1. Объединить данные таблиц tableau_project_data_1.csv, tableau_project_data_2.csv и tableau_project_data_3.csv, и изучить данные.
2. Построить дашборд "История выступлений", где будут изображены следующие графики:
    * Выступления по странам, где будут изображены все страны на карте, где происходили выступления и наиболее "активные" страны будут изображены темнее на карте;
    * Выступление по годам, учитывая фильтрацию по скачкообразному росту числа выступлений. Важно рассмотреть данные по топ-10 категориям выступлений как в относительном, так и в абсолютном выражении;
3. Построить дашборд "Тематики выступлений", где будут изображены следующие графики:
    * Отобразить топ-10 категорий по количеству выступлений. Также менее популярные, не входящие в топ-10 категории, запишем в отдельную категорию - "Другие" (Others);
    * Рассмотрим популярные категории в топ-3 популярных странах по количеству выступлений;
    * Также исследуем зависимость категории и длительность аплодисментов и смеха в секундах во время время выступления;
4. Построить дашборд "Авторы выступлений", где будут изображены следующие графики:
    * Наиболее популярный деятельности у авторов. Построим круговую диаграмму, где размер круга будет зависеть от количества выступлений в зависимости от деятельности авторов;
    * Посмотрим популярный авторов и их выступления по количеству просмотров;
5. Построить дашборд "Состав тематик категории "Другие"", где будут изображены следующие графики:
    * Распределение тематик по годам, которые находятся в категории "Другие";
    * Рассмотрим отдельно тематики 2019 г. и 2020 г. Посмотрим какие тематики были в 2019 и 2020 гг. и сравним.
